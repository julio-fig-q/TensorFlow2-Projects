# Okay lets build a Sequential model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense # a layer from keras; stands for densely connected
```

```python
# In this case, the model is a feed-forward neural net with a single layer

# We pass on a List of keras layers Dense(num_units, activation_func)
# Notice that there is no information about the input, weights, biases etc...
model = Sequential([Dense(64, activation = 'relu'),
                    Dense(10, activation='softmax')
                    ]) 

# However we can specify it from the beggining like so (this is optional) for first layer:
# And weights and biases will be initiated straight away; when input shape is specified
model = Sequential([
    Dense(64, activation='relu', input_shape=(784,)), # input is 1d vector of size 784
    Dense(10, activation='softmax')
])

# Another way of building the model is like such:

model = Sequential()

model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))

# And if the input is multidimensional it can be flattened using the Flatten layer:

model = Sequential([
    Flatten(input_shape = (28,28)), # (784,)
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
```
